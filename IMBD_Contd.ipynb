```python
df3 = mk.groupby('continent')[['population', 'gdp_cap']].aggregate(['max', 'min', 'count', 'mean'])
```

```python
# if we need max of population and average of gdp_cap column then we use dictonary
result = mk.groupby('continent')[['population', 'gdp_cap']].aggregate({'population' : 'max', 'gdp_cap' :'mean'})
```

```python
result.rename(columns = {'population':'pop_max', 'gdp_cap' : 'avg_gdp'})
```

```python
df4 = mk.groupby(['continent', 'year'])[['population', 'gdp_cap']].aggregate(['max', 'min'])
df4
```
